#### --> series


In pandas, a Series is a one-dimensional labeled array that can hold any data type (e.g., integers, floats, strings, etc.)
A Series consists of two main components: the data and the index. 
By default, the index is a sequence of integers starting from 0, but you can also specify your own custom index.

```python
cols2plot = ['fixed acidity','citric acid', 'quality'] # 3 columns to get from the main dataframe

# sns.jointplot is a function in the Seaborn data visualization library that creates a joint plot of two variables
# we want to explore the relationship between two variables and their distributions
# see more here: https://seaborn.pydata.org/generated/seaborn.jointplot.html

# we want to try : correlation of fixed acidity and citric acid : 
# Ä‘á»™ tÆ°Æ¡ng quan giá»¯a fixed acidity vÃ  citric acid thá»­ xem cÃ³ má»—i quan há»‡ gÃ¬ ko ....

sns.jointplot(
    # pass in a dataframe 
    # (here we limit the dataframe only have 3 RELEVANT columns, also helps speeding up the computation)
    data= data[cols2plot], 

    # 2 specific variables we want to check their relationship.
    x = 'fixed acidity', 
    y = 'citric acid',
    kind= 'reg' # show regressiion line 
)
```

```python
# looking in the above graph
# we saw the TOTAL DISTRIBUTION graph of all 'citric acid' in the outer right
# and the TOTAL DISTRIBUTION graph of all 'fixed acidity' on the outer top 
```

```python
# but since there are many Wine qualities, 
# we want to know how each quality performs 

cols2plot = ['fixed acidity','citric acid', 'quality'] # 3 columns to get from the main dataframe

sns.jointplot(
    # pass in a dataframe 
    # (here we limit the dataframe only have 3 RELEVANT columns, also helps speeding up the computation)
    data= data[cols2plot], 

    # 2 specific variables we want to check their relationship.
    x = 'fixed acidity', 
    y = 'citric acid',
    hue= 'quality',
    palette= 'tab10',
)
```

```python
# looking in the above graph
# we saw the DISTRIBUTION graph of 'citric acid' according to its 'quality' in the outer right
# and the DISTRIBUTION graph of 'fixed acidity' according to its 'quality' on the outer top 
```

```python
# pairwise plots

# but not always we are lucky enough to get linear pairs
# in real life: we just printed out bunch of combinations of pairs to see
# which graph has the best relationship

# --> the GOAL is to discover the relationship across different features of data that predict WINE's quality
# it's something that the model can discover on its own through training

# relevant columns
cols2plot = ['fixed acidity','volatile acidity','citric acid', 'quality']

# in pair plot, no variable is specified,
# it takes all columns of the dataframe we passed in (3 columns and quality as hue)
# to show their relationships
sns.pairplot(
    data[cols2plot], 
    hue='quality',
    palette= 'tab10',
    kind='reg' 
)
plt.show()
```

```python
# --- end code break down -------------------

# but looking at the above boxplot, we still have HUGE DIFFERENCE in the RANGE across different data features.
# so that's why we will Z-SCORE (standardize) each data feature individualy (except for QUALITY because that's our labels)

### z-score all variables except for quality

# find the columns we want to normalize (all except quality)
cols2zscore = data.keys()
cols2zscore = cols2zscore.drop('quality')

# z-score (written out for clarity)
for col in cols2zscore:
  meanval   = np.mean(data[col]) # mean value of data in current column
  stdev     = np.std(data[col],ddof=1) # standard deviation of data in current column
  data[col] = (data[col]-meanval) / stdev # finally the Z-SCORE (standardize) to resolve the HUGE DIFFERENCES in RANGE accoss different data features.

# can also do more compactly
# data[cols2zscore] = data[cols2zscore].apply(stats.zscore)

data.describe()
```

```python
# after stardization, we see that the MEAN value is almost 0 
# and stardard deviation is 1 all across
# which is the property of Z score graph.
```

```python
# check the plot again
fig,ax = plt.subplots(1,figsize=(17,4))
ax = sns.boxplot(data=data)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()
```
